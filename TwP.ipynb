```python
class Developer:
    def __init__(self,name,build_rate):
        self.name = name
        self.build_rate = build_rate
        self.build_rate_log = []
    
    def acquire(self, land, n_developers):
        return int(self.build_rate * (land / n_developers))
    
    def log_state(self):
        self.build_rate_log.append(self.build_rate)
```

```python
tiktok = Developer('TikTok-AI',0.3)
meta = Developer('Meta-AI',0.1)
google = Developer('Google-AI',0.05)
```

```python
print(tiktok.name)
print(tiktok.build_rate)
```

```python
class LandBank:
    def __init__(self, available_plots):
        self.available_plots = available_plots
        self.history = []
        self.available_plots_max = available_plots

    def develop(self, amount):
        self.available_plots = int(max(0,self.available_plots - amount))

    def log_state(self):
        self.history.append(self.available_plots)
    
    def release_plots(self,rate):
        self.available_plots = int(min(self.available_plots * (1 + rate), self.available_plots_max))
```

```python
land = LandBank(1000)
developers = [tiktok,meta,google]
n_developers=len(developers)
```

```python
for l in range(50):
    available_plots_acquireed = 0
    for ai in developers:
        available_plots_acquireed += ai.acquire(land.available_plots, len(developers))

    land.develop(available_plots_acquireed)
    land.release_plots(0.10)
    land.log_state()
    for i in developers:
        i.log_state()
```

```python
print(land.available_plots)
print(land.history)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.plot(range(50), land.history)
plt.title("Tragedia Wspólnego Pastwiska")
plt.xlabel("Rundy")
plt.ylabel("Zasób (kobiety w wieku rozrodczym)")
```
