```python
class algorithm:
    def __init__(self,name,aggressiveness):
        self.name = name
        self.aggressiveness = aggressiveness
        self.aggressiveness_log = []
    
    def harvest(self, resource):
        return self.aggressiveness * resource
    
    def log_state(self):
        self.aggressiveness_log.append(self.aggressiveness)
```

```python
tiktok = algorithm('TikTok-AI',0.9)
meta = algorithm('Meta-AI',0.5)
google = algorithm('Google-AI',0.3)
```

```python
print(tiktok.name)
print(tiktok.aggressiveness)
```

```python
class commonResource:
    def __init__(self, total):
        self.total = total
        self.history = []

    def deplete(self, amount):
        self.total = max(0,self.total - amount)

    def log_state(self):
        self.history.append(self.total)
```

```python
resource = commonResource(1000)
agents = [tiktok,meta,google]
```

```python
for l in range(50):
    total_harvested = 0
    for ai in agents:
        total_harvested += ai.harvest(resource.total)

    resource.deplete(total_harvested)
    resource.log_state()
    for i in agents:
        i.log_state()
```

```python
print(resource.total)
print(resource.history)
```
